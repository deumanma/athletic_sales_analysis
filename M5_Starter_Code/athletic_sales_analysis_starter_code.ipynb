```python
# Import Libraries and Dependencies
import pandas as pd
from pathlib import Path
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
#importing data

sales_2020_path = Path('Resources/athletic_sales_2020.csv') 
sales_2021_path = Path('Resources/athletic_sales_2021.csv')



# Read the CSV files into DataFrames.
sales_2020_df = pd.read_csv(sales_2020_path)
sales_2021_df = pd.read_csv(sales_2021_path)

```

```python
# Display the 2020 sales DataFrame
sales_2020_df.head()
```

```python
# Display the 2021 sales DataFrame
sales_2021_df.head()
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
sales_2020_df.dtypes
```

```python
# Check the 2021 sales data types.
sales_2021_df.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
# result = pd.concat([df1, df2], ignore_index=True)
# Example from class pd.concat([apple_data, goog_data, meta_data], axis="rows", join="inner"

sales_2020_2021_df = pd.concat([sales_2020_df, sales_2021_df], axis = "rows", join="inner", ignore_index = True)
sales_2020_2021_df
```

```python
# Check if any values are null.
sales_2020_2021_df.isnull()
```

```python
# Check the data type of each column
sales_2020_2021_df.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype

sales_2020_2021_df['invoice_date'] = pd.to_datetime(sales_2020_2021_df['invoice_date'], format='%Y-%m-%d')
sales_2020_2021_df.head()
```

```python
# Confirm that the "invoice_date" data type has been changed.
sales_2020_2021_df.dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.

product_sales_df = sales_2020_2021_df.groupby(['region', 'state', 'city'])['units_sold'].sum().reset_index()

# Rename the sum column to "Total_Products_Sold"
product_sales_df = product_sales_df.rename(columns={'units_sold': 'Total_Products_Sold'})

# Show the top 5 results.
group_top_products = product_sales_df.nlargest(5, 'Total_Products_Sold')
group_top_products                         
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.

pivot_product_sales_df = pd.pivot_table(sales_2020_2021_df, 
                             values='units_sold', 
                             index=['region', 'state', 'city'], 
                             aggfunc='sum').reset_index()

# Rename the "units_sold" column to "Total_Products_Sold"
pivot_product_sales_df = pivot_product_sales_df.rename(columns={'units_sold': 'Total_Products_Sold'})

# Show the top 5 results
pivot_top_products = pivot_product_sales_df.nlargest(5, 'Total_Products_Sold')
pivot_top_products
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.

group_total_sales = sales_2020_2021_df.groupby(['region', 'state', 'city'])['total_sales'].sum().reset_index()

# Rename the "total_sales" column to "Total Sales"
group_total_sales = group_total_sales.rename(columns={'total_sales': 'Total Sales'})

# Show the top 5 results
group_top_sales = group_total_sales.nlargest(5, 'Total Sales')
group_top_sales
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.

pivot_total_sales = pd.pivot_table(sales_2020_2021_df, 
                             values='total_sales', 
                             index=['region', 'state', 'city'], 
                             aggfunc='sum').reset_index()

# Optional: Rename the "total_sales" column to "Total Sales"
pivot_total_sales = pivot_total_sales.rename(columns={'total_sales': 'Total Sales'})

# Show the top 5 results
pivot_top_sales = pivot_total_sales.nlargest(5, 'Total Sales')
pivot_top_sales
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
women_athletic_footwear_sales = sales_2020_2021_df[sales_2020_2021_df['product'] == 'Women\'s Athletic Footwear']

# Setting the product as the index
women_athletic_footwear_sales.set_index('product', inplace=True)

# Display the result
women_athletic_footwear_sales.head()
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.

total_womens_footwear_sold_df = women_athletic_footwear_sales.groupby(['retailer', 'region', 'state', 'city'])['units_sold'].sum().reset_index()

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
total_womens_footwear_sold_df = total_womens_footwear_sold_df.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'})

# Show the top 5 results
group_top_womens_footwear = total_womens_footwear_sold_df.nlargest(5, 'Womens_Footwear_Units_Sold')
group_top_womens_footwear
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.

pivot_total_womens_footwear_sold_df = pd.pivot_table(women_athletic_footwear_sales, 
                             values='units_sold', 
                             index=['retailer', 'region', 'state', 'city'], 
                             aggfunc='sum').reset_index()

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
pivot_total_womens_footwear_sold_df = pivot_total_womens_footwear_sold_df.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'})

# Show the top 5 results
pivot_top_womens_footwear = pivot_total_womens_footwear_sold_df.nlargest(5, 'Womens_Footwear_Units_Sold')
pivot_top_womens_footwear
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.

pivot_max_sales_day = pd.pivot_table(sales_2020_2021_df, 
                                     values='units_sold', 
                                     index='invoice_date', 
                                     aggfunc='sum').reset_index()

# Optional: Rename the "units_sold" column to "Total Sales"
pivot_max_sales_day = pivot_max_sales_day.rename(columns={'units_sold': 'Total Sales'})

# Find the day with the maximum total sales
day_with_max_sales = pivot_max_sales_day.loc[pivot_max_sales_day['Total Sales'].idxmax()]
day_with_max_sales
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
bins_daily_sales = pivot_max_sales_day.resample('D', on='invoice_date').sum().reset_index()

# Sort the resampled pivot table in ascending order on "Total Sales"
sorted_daily_sales = bins_daily_sales.sort_values(by='Total Sales', ascending=False)
sorted_daily_sales
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.

bins_weekly_sales = pivot_max_sales_day.resample('W', on='invoice_date').sum().reset_index()

# Sort the resampled pivot table in ascending order on "Total Sales"
sorted_weekly_sales = bins_weekly_sales.sort_values(by='Total Sales', ascending=False)
sorted_weekly_sales
```
